```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
### Python dir() Function
### The dir() function returns all properties and methods of the specified object, 
### without the values.

dir(digits)
```

```python
digits.data[0]
```

```python
# Display an array as a matrix in a new figure window. 
# The origin is set at the upper left hand corner and rows 
# (first dimension of the array) are displayed horizontally. 
# The aspect ratio of the figure window is that of the array, 
# unless this would make an excessively short or narrow figure.

plt.gray()

# gray() Function. The gray() function in pyplot module of matplotlib library
# is used to set the colormap to “gray”.


for i in range( 5) :
    plt.matshow(digits.images[i])
```

```python
digits.target[0 : 5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data , digits.target , test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 10000)
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test , y_test)
```

```python
model.predict(X_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicted = model.predict(X_test)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test , y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm , annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

## now importing Logistics regression from sklearn.linear_model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 1000)
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cn = confusion_matrix(y_test , y_predicted)
cn
```

```python
import seaborn as sns
import matplotlib.pyplot as pl

pl.figure(figsize=(10, 7))
sns.heatmap(cn, annot=True)
pl.xlabel('Predicted')
pl.ylabel('Truth')
pl.show()
```
