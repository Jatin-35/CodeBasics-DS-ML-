```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_values =[1400,600,300,410,250]

exp_labels = ['Home Rent' , 'Food' , 'Phone Bill' , 'Car' , 'Other Utilities']
```

```python
# first argument as values ,  second arguement as labels

plt.axis('equal')
plt.pie(exp_values , labels = exp_labels)
plt.show()   # to remove unnecessary information
```

```python
# using radius , autopct (auto- percentage) , shadow (accepts boolean value)

plt.axis('equal')
plt.pie(exp_values , labels = exp_labels , radius = 2 , autopct = '%0.1f%%' , shadow = True  )
plt.show()   # to remove unnecessary information

```

```python
# using explode 


plt.axis('equal')
plt.pie(exp_values , labels = exp_labels , radius = 2 , autopct = '%0.1f%%' , shadow = True , explode = [0,0.2,0,0,0] )
plt.show()   # to remove unnecessary information
```

```python
# using startangle

plt.axis('equal')
plt.pie(exp_values , labels = exp_labels , radius = 2 , autopct = '%0.1f%%' , shadow = True , explode = [0,0.2,0,0,0] , startangle = 180 )
plt.show()   # to remove unnecessary information
```
