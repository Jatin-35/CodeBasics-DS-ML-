```python
import pandas as pd
df = pd.read_csv('salaries.csv')
df.head()
```

```python
inputs = df.drop('salary_more_then_100k' , axis = 'columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

### To add a column in the dataset


```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])

inputs.head()
```

### Fit_transform()
* It is used on the training data so that we can scale the training data and also learn the scaling parameters. Here, the model built will learn the mean and variance of the features of the training set. These learned parameters are then further used to scale our test data. *

```python
inputs_n = inputs.drop(['company','job' ,'degree'] , axis = 'columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n.values , target)
```

```python
model.score(inputs_n.values , target)
```

```python
model.predict([[2,2,1]])
```

```python
model.predict([[2,1,1]])
```
