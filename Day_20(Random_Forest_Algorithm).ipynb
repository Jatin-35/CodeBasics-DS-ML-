```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4) :
    plt.matshow(digits.images[i])
```

```python
plt.matshow(digits.images[25])
```

```python
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.drop(['target'], axis = 'columns') , df.target , test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 40)
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test , y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , y_predicted)
cm
```

```python
%matplotlib inline
import seaborn as sn
plt.figure (figsize = (10,7))
sn.heatmap(cm , annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.data
```

```python
iris.target
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data , columns = iris.feature_names )
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df.sample(10)
```

```python
df['target_names'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
```

```python
df.sample(10)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.drop(['target' , 'target_names'],axis = 'columns') , df['target' ] , test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier( n_estimators = 10)
model.fit(X_train,y_train)
```

```python

model.score(X_test , y_test)
```

```python
%matplotlib inline
score = []
n_estimators =  range(1,300)
for i in n_estimators :
    model = RandomForestClassifier( n_estimators = i)
    model.fit(X_train,y_train)
    score.append(model.score(X_test , y_test) )
    print('index : {} , score:{} , n_estimator :{}'.format(i-1 ,score[i-1],i))
    
plt.plot(score,n_estimators)
plt.xlabel('n_estimators')
plt.ylabel('testing accuracy')
 
    
```

```python
maximum_score = max(score)
maximum_score
```

```python

score.index(maximum_score)
```
