```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6]
max_t = [23,45,67,48,22,33]
min_t = [35,43,13,32,43,24]
avg_t = [30,56,76,33,45,48]
```

```python

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Stats")
plt.plot(days,max_t, label = "Max")
plt.plot(days,min_t ,  label = "Min")
plt.plot(days,avg_t ,  label = "Avg")

plt.legend(loc = "best" , shadow = "True" , fontsize = "large")

plt.grid()

#'Best' is not a valid value for loc; supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right',
#'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
```

## Bar Graph

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
company = ['GOOGLE' , 'AMAZON' , 'APPLE' , 'SAMSUNG']
revenue = [55,34,53,34]
profit = [23,45,56,73,]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos , company)
plt.bar(xpos -0.2 , revenue,  width = 0.4)
plt.bar(xpos +0.2,profit , width = 0.4)
plt.xlabel("Companies")
plt.ylabel("Revenue")
plt.title('Stats')
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.yticks(ypos , company)
plt.barh(ypos -0.2 , revenue , height = 0.4)
plt.barh(ypos +0.2,profit , height = 0.4 )
plt.xlabel("Companies")
plt.ylabel("Revenue")
plt.title('Stats')
```
