# Dealing with Missing Data

## using fillna() , dropna() , interpolate()

```python
import pandas as pd
```

```python
df = pd.read_csv('weather1_data.csv')
df
```

```python
df = pd.read_csv('weather1_data.csv')
type(df.day[0])
```

```python
# we need to set day as index so first we convert it into timestamp datatype by using parse_dates

df = pd.read_csv('weather1_data.csv' , parse_dates = ['day'] )
type(df.day[0])
```

```python

# we need to set the index as day for that we will use the set_index()

df = pd.read_csv('weather1_data.csv' , parse_dates = ['day'] )
df.set_index('day' , inplace = True)
df
```

### Now lets deal with the missing data : 

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna(
{
    'temperature' : 0,
    'windspeed'  : 0,
    'event' : 'no event'
    
    
})

new_df
```

```python
new_df = df.fillna(method = 'ffill')    ## ffills = forward fill
new_df 
```

```python
new_df = df.fillna(method = 'bfill')     ## bfills = backward fill 
new_df 
```

```python
new_df = df.fillna(method = 'bfill' , limit = 1)     ## limit = how many cells next to filled cell will have the value
new_df 
```

```python
new_df = df.fillna(method = 'bfill' , axis = 'columns')     ## xis : how the value to be filled inthis case columnwise
new_df 
```

```python
new_df = df.interpolate()    
new_df
```

```python
new_df = df.interpolate(method = 'quadratic')    
new_df
```

### Method must be one of :
                       ['linear', 'time', 'index', 'values', 'nearest', 'zero', 'slinear', 
                       'quadratic', 'cubic', 'barycentric', 'krogh', 'spline', 'polynomial', 'from_derivatives', 
                       'piecewise_polynomial', 'pchip', 'akima', 'cubicspline']. Got 'Time' instead.

```python
new_df = df.interpolate(method = 'time')    
new_df
```

```python
new_df = df.dropna()    
new_df
```

```python
new_df = df.dropna(how = 'all')     # all means if all columns have all NaN values it will be dropped
new_df
```

```python
new_df = df.dropna(thresh = 1)    
new_df
```

```python
new_df = df.dropna(thresh = 2)    # threshhold value
new_df
```
